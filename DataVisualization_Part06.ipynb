## <font color='blue'> Description </font>

> We will model the data from the file 'GoalsInLeague.txt'. This file contains the results of a number of football matches from a certain football league. The first line of the file stores the number of goals scored by the home teams and the second line stores the number of goals scored by the visiting teams.

## <font color='blue'> Loading the Required Packages </font>

```python
import matplotlib.pyplot as plt                    
from matplotlib.ticker import MultipleLocator      
from matplotlib.ticker import FormatStrFormatter
from matplotlib.patches import Rectangle
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import math
import warnings
warnings.filterwarnings('ignore')
#%matplotlib widget
```

## <font color='blue'> Opening, Reading, Formatting the File and Storing Data in Lists </font> 

```python
with open("GoalsInLeague.txt", "r") as dataGoals:
    # First line
    TheHomeTeam = dataGoals.readline()
    TheHomeTeam = TheHomeTeam.strip(" \n") # Extra space + \n
    TheHomeTeam = TheHomeTeam.split(" ")

    TheHomeTeamGoals = [int(x) for x in TheHomeTeam]

    # Second Line
    TheVisitingTeam = dataGoals.readline()
    TheVisitingTeam = TheVisitingTeam.strip(" \n")
    TheVisitingTeam = TheVisitingTeam.split(" ")

    TheVisitingTeamGoals = [int(x) for x in TheVisitingTeam]
```

```python
#print(TheHomeTeamGoals)
#print(TheVisitingTeamGoals)
```

## <font color='blue'> Pie chart </font> 

```python
TotalHomeTeamGoals = sum(TheHomeTeamGoals)
TotalVisitingTeamGoals = sum(TheVisitingTeamGoals) 
```

```python
# Plot area definition.
fig = plt.figure("Third", figsize = (10, 10)) #(width, height) 

ax1 = fig.add_subplot(2, 2, 1) 
ax1.pie(x = [TotalHomeTeamGoals, TotalVisitingTeamGoals],
        labels = ['Home Team', 'Visiting Team'],
        labeldistance = 1.2,
        colors = ['red', 'black'],
        explode = [0, 0.1],
        radius = 1.6)

ax2 = fig.add_subplot(2, 2, 2) 
ax2.pie(x = [0.2, 0.3],
        # sum(x) > 1 -> x[i]/sum(x) 
        # sum(x) <= 1 -> there is no normalization 
        labels = ['Home Team', 'Visiting Team'],
        labeldistance = 0.2,
        colors = ['yellow', 'w'],
        explode = [0.1, 0],
        shadow = True,
        radius = 1.2)
 
ax3 = fig.add_subplot(2, 2, 3) 
ax3.pie(x = [0.3, 0.4, 0.4],
        labels = ['Home Team', 'Visiting Team', 'Rest'],
        colors = ['pink', 'b', 'purple'],
        explode = [0.1, 0, 0],
        shadow = True)

ax4 = fig.add_subplot(2, 2, 4) 
ax4.pie(x = [0.3, 0.4, 0.4, 5],
        labels = ['Home Team', 'Visiting Team', 'Rest1', 'Rest2'],
        radius = 0.8)

fig.tight_layout(pad = 4.0)

plt.show()
```

```python
# Plot area definition.
fig1 = plt.figure("Fourth", figsize = (4, 4)) #(width, height) 

ax5 = fig1.add_subplot(1, 1, 1) 
ax5.pie(x = [TotalHomeTeamGoals, TotalVisitingTeamGoals],
        labels = ['Home Team', 'Visiting Team'],
        labeldistance = 1.2,
        colors = ['red', 'brown'],
        explode = [0, 0.1],
        radius = 1.0,
        center = [1, 1])

ax5.pie(x = [TotalHomeTeamGoals, TotalVisitingTeamGoals],
        labels = ['Home Team', 'Visiting Team'],
        labeldistance = 1.2,
        colors = ['red', 'brown'],
        explode = [0, 0.1],
        radius = 1.5,
        center = [-2, -2]) # center = Chart location.

plt.show()
```
